<p>Este trabalho foi-me proposto pelo Professor José Carlos Ramalho, no âmbito da cadeira de Algoritmos e Técnicas de Programação. Com isto, pretendia-se que fosse elaborado um menu que permitisse a realização de várias operações envolvendo polinómios.</p>
<p>Durante a aula da cadeira, foram elaboradas e explicadas as seguintes funções:"criarTermo", "criarPolinomio", "inserirTermo", "verTermo", "calcularPolinomio", "tabela", "chaveOrd", "ordenarPol" e "derivarPolinomio".</p>
<p>Assim sendo, na função "simplificarPol", começamos por ordenar o polinómio para depois ser mais fácil observar quais os termos com o mesmo expoente. De seguida, criei um novo polinómio que, no inicio, consiste apenas numa lista vazia. Para além disso, acrescentei um ciclo for que vai percorrer todos os termos do nosso polinómio, que ainda constituem uma lista de tuplos. Como tal, criei um termo t1, com o coeficiente e o expoente do polinómio p, mas que em vez de ser um tuplo é agora uma lista, lista essa que vai ser adicionada ao p1, que anteriormente estava vazio. Assim, p1 vai ser uma lista de listas. Seguidamente, inseri um ciclo while que vai ser válido enquanto uma variável i for menor do que o comprimento do p1-1, isto porque o último índice não dá para comparar pois não tem mais nenhum a seguir a ele. Dentro deste ciclo, tem-se uma condição que compara os valores dos expoentes, ou seja, apenas vai ser válida para termos que tenham igual expoente. Surge, então, a variável c que é a soma dos dois coeficientes caso sejam iguais os expoentes, criando assim depois o termos t que resulta da soma dos dois coeficientes, elevados ao expoente que tinham. Como tal, esse termo é inserido no polinomio p1. Por outro lado, como estamos a comparar dois termos de cada vez, quando aparecem dois com o mesmo expoente esperasse que surja apenas um único termo resultante desses dois. Para isso, usa-se a função pop duas vezes correspondente aos dois termos de igual expoente para os retirar do polinómio. Ao procedermos assim, vamos ficando com menos termos no polinómio, havendo depois no final a necessidade de o voltar a ordenar. No fim iguala-se o i a 0 para que quando houver termos de igual expoente se possa proceder ao ciclo novamente. Caso, não tenha o mesmo expoente ele vai fazer i=i+1 e assim vão sendo comparados os termos dois a dois, até não ocorrer mais alterações e termos o retorno do polinómio simplificado.</p>
<p> Por útimo, temos a função "fazer" que permite pôr em prática as opções do menu. É nesta função que se pede ao utilizador os diferentes valores necessários para a realização das operações e a que evoca as diversas funções do código conforme a necessidade de cada opção.</p>

```python
def menu():
    print(""" Menu de Opções
(1) Criar polinómio
(2) Calcular Polinómio
(3) Calcular Tabela
(4) Simplificar Fração
(5) Derivar polinómio
(0) Sair """)


def criarTermo(c, e):
    return(c, e)

def criarPolinomio():
    return([])

def inserirTermo(p, t):
    return(p.append(t))

def verTermo(t):
    c, e = t
    if e == 0:
        return(str(c))
    else:
        return(str(c) + 'x^' + str(e))


def verPolinomio(p):
    res = " "
    for t in p:
        if res == " ":
            res = verTermo(t)
        else:
            res = res + "+" + verTermo(t)
    print(res)

def calcularPolinomio(p, x):
    res = 0
    for (c, e) in p:
        res = res + c *(x ** e)
    return(res)

def tabela(p, nlinhas):
    for i in range(nlinhas):
        print(str(i) + "::" + str(calcularPolinomio(p, i)))


def chaveOrd(t):
    c, e = t
    return(e)

def ordenarPol(p):
    p.sort(reverse = True, key = chaveOrd)
    return(p)

def simplificarPol(p):
    p = ordenarPol(p)
    p1 = criarPolinomio()
    for t in p:
        t1=[t[0],t[1]]
        inserirTermo(p1,t1)
    i=0
    while i<len(p1)-1:
        if p1[i][1]==p1[i+1][1]:
            c=p1[i][0]+p1[i+1][0]
            t=criarTermo(c,p1[i][1])
            inserirTermo(p1,t)
            p1.pop(i)
            p1.pop(i)
            p1=ordenarPol(p1)
            i=0
        else:
            i=i+1
    return(p1)

def derivarPolinomio(p):
    res = []
    for t in p:
        c, e = t
        if e > 0:
            res.append((c*e, e-1))
    return(res)



def fazer():
    opcao = "1"
    menu()
    while opcao != "0":
        opcao = input("Qual a sua opção? ")
        if opcao == "1":
            polinomio = int(input("Quantos termos tem o polinomio? "))
            p = criarPolinomio()
            for i in range(polinomio):
                c = int(input("Qual o valor do coeficiente? "))
                e = int(input("Qual o valor do expoente? "))
                t = criarTermo(c, e)
                a = inserirTermo(p, t)
            verPolinomio(p)
        elif opcao == "2":
            polinomio = int(input("Quantos termos tem o polinomio? "))
            p = criarPolinomio()
            for i in range(polinomio):
                c = int(input("Qual o valor do coeficiente? "))
                e = int(input("Qual o valor do expoente? "))
                t = criarTermo(c, e)
                a = inserirTermo(p, t)
            x = int(input("Qual o valor de x? "))
            a = calcularPolinomio(p, x)
            print(a)
        elif opcao == "3":
            nlinhas = int(input("Qual o número de linhas da tabela? "))
            tabela(p, nlinhas)
        elif opcao == "4":
            polinomio = int(input("Quantos termos tem o polinomio? "))
            p = criarPolinomio()
            for i in range(polinomio):
                c = int(input("Qual o valor do coeficiente? "))
                e = int(input("Qual o valor do expoente? "))
                t = criarTermo(c, e)
                a = inserirTermo(p, t)
            a = simplificarPol(p)
            verPolinomio(a)
        elif opcao == "5":
            polinomio = int(input("Quantos termos tem o polinomio? "))
            p = criarPolinomio()
            for i in range(polinomio):
                c = int(input("Qual o valor do coeficiente? "))
                e = int(input("Qual o valor do expoente? "))
                t = criarTermo(c, e)
                a = inserirTermo(p, t)
            p = simplificarPol(p)
            a = derivarPolinomio(p)
            verPolinomio(a)
        elif opcao == "0":
            print("Até à próxima...")
        else:
            print("Opção inválida")
print(fazer())
            
```
