### Resumo

<p>Este trabalho foi-me proposto pelo Professor José Carlos Ramalho, no âmbito da cadeira de Algoritmos e Técnicas de Programação. Com isto, pretendia-se que fosse elaborado um código onde o computador, num máximo de 7 tentativas, descobrisse qual o número que uma determinada pessoa pensou.</p>
	<p>Como tal, comecei por definir uma função onde se começa com 0 tentativas, isto é, até ao momento que o computador apresentar um valor aleatório, o número de tentativas efetuadas para descobrir o valor real é 0. De seguida, apresentei uma string, cujo principal objetivo é explicar ao jogador que deve pensar num número que se encontre entre 0 e 100. Na linha seguinte, designei por 'encontrado' o valor real, atribuindo-lhe o valor falso pois quando o jogo se inicia o computador ainda não sabe de que valor se trata. Posteriormente, criei duas variàveis às quais dei o valor de 50, na medida em que numa escala de 0 a 100, a melhor hipótese de o computador advinhar o número real em menores tentativas é começar por um número que tenha 50/50 possibilidade de ser o correto. A diferença entre estas duas variáveis é que a variável b é o ponto de partida (b=50) para que fosse a base da soma ou subtração, num caso de tentativas superior a 1. Por sua vez, a variável c é aquela que é obtida através das sucessivas divisões e a qual é retirada ou acrescentada à variável b conforme as diferentes situações.</p>
	<p>Assim, para que o jogo se faça de forma repetitiva até ser encontrado o valor, criei um ciclo while, estabelecendo a condição de que o ciclo se vai repetir apenas enquanto o 'encontrado' não tiver associado o valor verdadeiro. Então, forcei o computador a considerar que o melhor ponto de partida é dizer que o valor certo é 50, pois tem metade de possibilidades de estar na metade dos valores superiores e metade de possiblidades de estar na metade dos valores inferiores. Para além disso, criei outra string que pretende ajudar o computador a saber se o valor que ele atribui aleatoriamente está acima, abaixo ou se é o valor real.</p>
	<p>Ora, no caso de o valor verdadeiro ser maior do que aquele selecionado aleatoriamente pelo computador, este deve proceder uma vez mais à divisão por dois, para que possa ficar outra vez com 50/50 de possibilidades de estabelecer um intervalo menor de valores. Ao valor obtido através dessa divisão é somado o valor da variável b, pois este é o ponto de partida do computador. Feita uma tentativa do computador, é somado 1 ao valor de tentativas inicial, que depois é apresentado.</p>
	<p>Por outro lado, quando o valor verdadeiro é menor do que aquele selecionado pelo computador, procede-se novamente à divisão por dois de forma a cada vez mais diminuir o número de hipóteses até se encontrar o valor. Contudo, neste caso subtrai-se este valor à variável b. Também aqui é somado 1 ao número de tentativas inicial, pois pretende-se ficar a saber qual o número máximo de tentativas que foram necessárias até se chegar ao valor.</p>
	<p>Por último, no caso de o valor verdadeiro corresponder ao dito pelo computador, a variável 'encontrado' assume um valor verdadeiro. Ao assumir tal, o ciclo while termina e soma-se novamente 1 ao número de tentativas. No final aparece o número total das tentativas usadas e o valor que o jogador pensou.</p>
