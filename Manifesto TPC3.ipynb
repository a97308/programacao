### Resumo

<p>Este trabalho foi-me proposto pelo Professor José Carlos Ramalho, no âmbito da cadeira de Algoritmos e Técnicas de Programação. Pretendia-se criar um menu com diferentes opções, nas quais se destacava a função bubble sort.</p>
<p> As funções "menu", "criarlistaA", "criarlistaM" e "maior", foram abordadas e explicadas durante a aula teórica. Como tal, restava elaborar a função bubble sort.</p>
<p> Assim, para a opção 4 do menu, comecei por evocar a função "criarlistaA", a qual designei por "lista2". De seguida, igualei i=0, pois quando começamos a correr a lista, tal se inicia na posição 0. Para além disso, designei uma outra variável por "alteracao", dando-lhe o valor 1, para depois poder iniciar um ciclo while no qual a variável "alteracao" é diferente de 0. Já dentro do ciclo, posso então dizer que a variável assume, agora, o valor 0 para iniciar. Dentro deste ciclo, crio um outro ciclo for, no qual a variável i tem um alcance do comprimento da lista - 1. Isto porque como os elementos são comparados dois a dois, para n elementos temos n-1 passos. Dentro do ciclo crio uma condição para quando um número à esquerda na lista é maior do que um número à direita. Neste caso, há uma troca de posições, soma-se 1 às alterações feitas e 1 ao i pois avança uma posição. Caso se verifique o contrário, apenas soma-se 1 à posição, havendo posteriormente um returno da lista com os elementos todos ordenados.</p>
