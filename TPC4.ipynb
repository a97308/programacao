<p>Este trabalho foi-me proposto pelo Professor José Carlos Ramalho, no âmbito da cadeira de Algoritmos e Técnicas de Programação. Com isto, pretendia-se que fosse elaborado um menu onde se apresentasse várias opções que podem ser concretizáveis com frações.</p>
<p> Durante a aula da cadeira, foram elaboradas e explicadas as funções: "criarFracao", correspondente à opção 1; "verFracao", que embora não corresponda a nenhuma opção, é esta a função que permite a visualização dos resultados; "mdc", que é uma função auxiliadora noutra opção; "simplificarFracao", que corresponde à opção 2; "somarFracao", que corresponde à opção 3; e "somalista", correspondente à opção 8. Como tal, restavam fazer 6 funções. </p>
<p> Assim, primeiramente foi elaborada a função menu que permite ao utilizador consultar as opções disponíveis. De seguida, criei as funções "subtrairFracao", "multiplicarFracao" e "divFracao", correspondentes, respetivamente, às opções 4, 5 e 6, nas quais defini do denominador e numerador das 2 frações, apresentado depois a respetiva expressão matemática para o cálculo.</p>
<p>Para além disso, no que diz respeito à opção 7, defini a função "criarlista", começando por criar uma lista aberta. Seguidamente, disse que para um número máximo n, definido pelo utilizador, o denominador e numerador seriam um número entre 1 e 20 e que cada fração criada, aleatoriamente, seria então acrescentada à lista inicialmente vazia. Para que a lista pudesse estar definida, na função "fazer" na parte correspondente à opção 7, acrescentei uma string onde o utilizador pode, assim, atribuir um valor a n, de modo que possa depois ser criada uma lista.</p>
<p>No que diz respeito à opção 9, defini a função "maiorFracao", começando por definir duas variáveis. A primeira variável é um tuplo, onde o primeiro [0] se refere ao primeiro tuplo e o segundo [0] se refere à primeira parte desse mesmo tuplo. Por outro lado, a segunda variável corresponde à divisão da primeira parte do primeiro tuplo pela segunda parte do primeiro tuplo. Esta divisão é necessária, pois o seu resultado é que permite saber qual a maior fração. Como tal, para um tuplo i que esteja dentro dos limites da lista, acrescentei uma variável, à qual lhe dei o valor resultante da divisão da primeira parte de um dado tuplo i pela segunda parte desse mesmo tuplo. Assim, caso o valor dessa variável for superior ao da variável "maior2", esta última assume o valor da variável "div" e, consequentemente, depois de percorrida a lista, a variável "maior1" vai então assumir o valor da maior fração.</p>
<p>Por último, a função "fazer" permite pôr o menu em prática, sendo que para cada opção selecionada tem sempre uma string a perguntar ao utilizador os valores que este pretenda inserir.</p>

#### Código:

```python
import random

def menu():
    print("""Menu de opções
(1) Criar fração;
(2) Simplificar fração;
(3) Somar frações;
(4) Subtrair frações;
(5) Multiplicar frações;
(6) Dividir frações;
(7) Gerar lista de n frações;
(8) Somar uma lista;
(9) Maior fração da lista;
(0) Sair""")



def criarFracao(numerador, denominador):
    return(numerador, denominador)

def verFracao(f):
    print(str(int(f[0])) + " / " + str(int(f[1])))


def mdc(a,b):
    if a < b:
        return(mdc(b, a))
    elif (a%b) == 0:
        return(b)
    else:
        return (mdc(a, a%b))

def simplificarFracao(f):
    numerador, denominador = f
    a = mdc(numerador, denominador)
    return(numerador/a, denominador/a)

def somarFracao(f1, f2):
    numerador1, denominador1 = f1
    numerador2, denominador2 = f2
    return(simplificarFracao(criarFracao(numerador1*denominador2 + numerador2*denominador1, denominador1*denominador2)))

def subtrairFracao(f1, f2):
    numerador1, denominador1 = f1
    numerador2, denominador2 = f2
    return(simplificarFracao(criarFracao(numerador1*denominador2 - numerador2*denominador1, denominador1*denominador2)))


def multiplicarFracao(f1, f2):
    numerador1, denominador1 = f1
    numerador2, denominador2 = f2
    return(simplificarFracao(criarFracao(numerador1*numerador2, denominador1*denominador2)))

def divFracao(f1, f2):
    numerador1, denominador1 = f1
    numerador2, denominador2 = f2
    return(simplificarFracao(criarFracao(numerador1*denominador2, denominador1*numerador2)))

def criarlista(n):
    lista = []
    for i in range(n):
        numerador = random.randint(1, 20)
        denominador = random.randint(1, 20)
        f = (numerador, denominador)
        lista.append(f)
    return(lista)

def somalista(lista):
    res = (0, 1)
    for fracao in range(len(lista)):
        res = somarFracao(lista[fracao], res)
    verFracao(res)

def maiorFracao(lista):
    maior1 = (lista[0][0],lista[0][1])
    maior2 = (lista[0][0]/lista[0][1])
    for i in range(len(lista)):
        div = (lista[i][0])/(lista[i][1])
        if div > maior2:
            maior2 = (lista[i][0]/lista[i][1])
            maior1 = (lista[i][0],lista[i][1])
    return(maior1)
    
    

def fazer():
    opcao = "1"
    menu()
    while opcao != "0":
        opcao = input("Qual a sua opção?")
        if opcao == "1":
            numerador = int(input("Qual é o numerador: "))
            denominador = int(input("Qual o denominador: "))
            a = criarFracao(numerador, denominador)
            print(a)
        elif opcao == "2":
            numerador = int(input("Qual é o numerador: "))
            denominador = int(input("Qual o denominador: ")) 
            f = (numerador, denominador)
            a = verFracao(simplificarFracao(f))
            print(a)
        elif opcao == "3":
            numerador1 = int(input("Qual é o numerador: "))
            denominador1 = int(input("Qual o denominador: "))
            f1 = criarFracao(numerador1, denominador1)
            numerador2 = int(input("Qual é o numerador: "))
            denominador2 = int(input("Qual o denominador: "))
            f2 = criarFracao(numerador2, denominador2)
            a = verFracao(somarFracao(f1, f2))
            print(a)
        elif opcao == "4":
            numerador1 = int(input("Qual é o numerador: "))
            denominador1 = int(input("Qual o denominador: "))
            f1 = criarFracao(numerador1, denominador1)
            numerador2 = int(input("Qual é o numerador: "))
            denominador2 = int(input("Qual o denominador: "))
            f2 = criarFracao(numerador2, denominador2)
            a = verFracao(subtrairFracao(f1, f2))
            print(a)
        elif opcao == "5":
            numerador1 = int(input("Qual é o numerador: "))
            denominador1 = int(input("Qual o denominador: "))
            f1 = criarFracao(numerador1, denominador1)
            numerador2 = int(input("Qual é o numerador: "))
            denominador2 = int(input("Qual o denominador: "))
            f2 = criarFracao(numerador2, denominador2)
            a = verFracao(multiplicarFracao(f1, f2))
            print(a) 
        elif opcao == "6":
            numerador1 = int(input("Qual é o numerador: "))
            denominador1 = int(input("Qual o denominador: "))
            f1 = criarFracao(numerador1, denominador1)
            numerador2 = int(input("Qual é o numerador: "))
            denominador2 = int(input("Qual o denominador: "))
            f2 = criarFracao(numerador2, denominador2)
            a = verFracao(divFracao(f1, f2))
            print(a) 
        elif opcao == "7":
            n = int(input("Introduza o nº de frações da lista: "))
            a = verFracao(criarlista(n))
            print(a)
        elif opcao == "8":
            n = int(input("Introduza o nº de frações da lista: "))
            lista = criarlista(n)
            print(lista)
            a = somalista(lista)
            print(a)
        elif opcao == "9":
            n = int(input("Introduza o nº de frações da lista: "))
            lista = criarlista(n)
            print(lista)
            a = verFracao(maiorFracao(lista))
            print(a)
        elif opcao == "0":
            print("Até à próxima...")
        else:
            print("Opção inválida")
print(fazer())
```
